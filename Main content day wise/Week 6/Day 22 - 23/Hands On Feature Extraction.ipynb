```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
path = "C:\\Users\\devid\\Desktop\\datasets\\Automobile_data.csv"

# C:\Users\devid\Desktop\datasets\Automobile_data.csv"          ^
# SyntaxError: (unicode error) 'unicodeescape' codec can't decode 

df = pd.read_csv(path)
df.head()
```

```python
# Handling missing values (replace '?' with NaN and drop missing values)
df.replace("?", np.nan, inplace=True)  # replaces all occurrences of "?" in the df with NaN
df.dropna(inplace=True) # removes any rows in the DataFrame that contain NaN values

# Converting necessary columns to numeric
for col in ["normalized-losses", "bore", "stroke", "horsepower", "peak-rpm", "price"]:
    df[col] = pd.to_numeric(df[col])
```

```python
df
```

**Step 2: Polynomial Features (Numerical Data)**


You can generate polynomial features for numerical columns such as curb-weight, engine-size, and horsepower to capture non-linear relationships between them.

```python
from sklearn.preprocessing import PolynomialFeatures

# Select numerical columns
numerical_features = df[['curb-weight', 'engine-size', 'horsepower']]
```

```python
# Create polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
poly_features = poly.fit_transform(numerical_features)
```

`PolynomialFeatures(degree=2)`:

This initializes a PolynomialFeatures object with a degree of 2.

`poly.fit_transform(numerical_features)`

learns the transformation based on the data, e.g. calculates how to create the higher-degree terms (like x^2, y^2, x*y).

After fitting, this method applies the transformation to the input data (numerical_features).


```python
poly.get_feature_names_out(numerical_features.columns)
```

```python


# Create a DataFrame of new features
poly_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(numerical_features.columns))

# Display the original and new polynomial features
print("Original Features and Polynomial Features:")
print(poly_df.head())
```

```python
poly_df.head()
```

**Step 3: Interaction Features (Numerical Data)**

You can create interaction features by combining two numerical features, such as multiplying engine-size and curb-weight to capture interaction effects between the engine and car weight.

```python
# Create interaction feature between engine-size and curb-weight
df['engine_curb_weight_interaction'] = df['engine-size'] * df['curb-weight']

# Display the new feature
print("Interaction Feature (engine-size * curb-weight):")
print(df[['engine-size', 'curb-weight', 'engine_curb_weight_interaction']].head())
```

```python
df.head()
```

**Step 4: Binning (Numerical to Categorical)**

We can convert continuous numerical variables into categorical ones. For example, binning the horsepower into categories: Low, Medium, and High.

```python
# Create bins for horsepower
bins = [0, 100, 150, 200]
labels = ['Low', 'Medium', 'High']

df['horsepower_binned'] = pd.cut(df['horsepower'], bins=bins, labels=labels)
```

`pd.cut()` is a function in pandas that is used to segment and sort data into discrete bins. It takes continuous values and categorizes them into predefined intervals (bins).

```python


# Display the binned feature
print("Binned Feature (horsepower):")
print(df[['horsepower', 'horsepower_binned']].head())
```

**Step 5: Encoding Categorical Features (Textual Data)**

You can encode categorical columns like body-style using one-hot encoding or label encoding.

```python
# One-hot encode the 'fuel-type' column
df_onehot = pd.get_dummies(df, columns=['body-style'], drop_first=True)
df_onehot.head()
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder

# Label encode the 'body-style' column
le = LabelEncoder()
df['body-style_encoded'] = le.fit_transform(df['body-style'])

# Display the label-encoded feature
print("Label Encoded Feature (body-style):")
print(df[['body-style', 'body-style_encoded']].head())
```

for `label_encoders = {}` The keys will be the column names of your DataFrame df that contain categorical data (i.e., columns of type 'object'), and the values will be the LabelEncoder objects used to encode these categorical values into numerical values.

```python
# Encoding categorical columns

label_encoders = {}

for column in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

df.head()
```

### **Step 6: Dimensionality Reduction (Optional)**

Using PCA to reduce the dimensionality of a feature set might be useful if the dataset has a lot of numerical features. For simplicity, we'll use some of the numerical features.


```python
from sklearn.decomposition import PCA

```

```python


# Select numerical features for PCA
numerical_features = df[['curb-weight', 'engine-size', 'horsepower', 'city-mpg', 'highway-mpg']]

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions

pca_features = pca.fit_transform(numerical_features)

# Display the reduced features
print("Reduced Features using PCA:")
print(pca_features[:5])
```

The Output `(pca_features[:5])`:

- The output shows the first 5 rows of the transformed data. 
- Each row represents a car's data projected onto the 2 principal components.
- These two values represent the coordinates of each data point (car) in the new 2-dimensional space defined by the principal components.

In very simple terms, after applying PCA to your data, you transformed your original 5 features (like weight, engine size, horsepower, etc.) into just 2 new features called principal components.

Instead of thinking about 5 attributes for each car, you're now thinking about 2 combined values that capture the most important differences between the cars.

The first value in each pair tells you how the car is positioned along the first principal component (which captures the biggest difference between cars).

The second value tells you how the car is positioned along the second principal component (which captures the next biggest difference).

```python
df.head()
```

```python
# Convert PCA features to a DataFrame with appropriate column names
pca_df = pd.DataFrame(pca_features, columns=['PCA1', 'PCA2'])
```

```python
# Add the PCA features back to the original dataframe
df = pd.concat([df, pca_df], axis=1)
```

```python
df.head()
```

```python
# Drop the original numerical columns to reduce complexity
df.drop(columns=['curb-weight', 'engine-size', 'horsepower', 'city-mpg', 'highway-mpg'], inplace=True)

```

```python
# Display the updated dataframe
df.head()
```

Apply PCA again on the updated dataset (which now includes PCA1 and PCA2)

```python
# Select the columns to apply PCA again (excluding non-numerical columns)
columns_for_pca = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
columns_for_pca
```

```python
# Impute missing values with the median of each column
df[columns_for_pca] = df[columns_for_pca].fillna(df[columns_for_pca].median())
```

```python
# Apply PCA again on the dataset
pca = PCA(n_components=2)  # Reduce to 2 dimensions again
pca_features = pca.fit_transform(df[columns_for_pca])
```

```python
# Convert PCA features to a DataFrame with appropriate column names
pca_df = pd.DataFrame(pca_features, columns=['PCA1_reduced', 'PCA2_reduced'])
```

```python
# Add the new PCA features back to the original dataframe
df = pd.concat([df, pca_df], axis=1)
```

```python
# Drop the original columns that were used for the second PCA
df.drop(columns=columns_for_pca, inplace=True)
```

```python
# Display the updated dataframe with reduced columns
df.head()
```
