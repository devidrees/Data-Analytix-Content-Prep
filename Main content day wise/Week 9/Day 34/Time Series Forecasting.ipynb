    <h4> 1. ARIMA (AutoRegressive Integrated Moving Average)
    </h4> </div>

  ARIMA is a popular and powerful statistical model for time series forecasting. 
  
  It is particularly effective for data that shows trends but lacks strong seasonal components.  


**Imagine** you're guessing tomorrow's weather based on how similar today is to yesterday and the day before.

- AR looks at past values (lags) and finds a relationship.
- Example: 
    - If it has been sunny for the past two days, there's a good chance tomorrow will be sunny as well.
    - If ice cream sales were high for the past two Saturdays (hot weekends), we might predict high sales this coming Saturday as well.

**Components of ARIMA:**  
   ARIMA consists of three main parts, denoted as **(p, d, q):**  
  
   **1. Autoregressive (AR):**  
   - "Use the Past to Predict the Present"
   - Uses the relationship between an observation and a specified number of lagged observations.  
   
   - **Parameter (p):** Determines the number of lagged terms to include.  
   - **Example:** Predicting today's sales based on the last two daysâ€™ sales. 

 **2. Integrated (I):**  
- Refers to differencing the data to achieve stationarity (removing trendsand seasonality). 
-  The "Integrated" part **removes trends** to make the data stationary (nolong-term rise/fall).
 
- **Parameter (d):** Specifies the number of differences needed to make theseries stationary.  
    - **Example:** Converting a time series from cumulative sales to dailysales by differencing (subtracting one value from the next).
        - Sales: [100, 200, 300, 400] - Before Differencing
        - Sales Change: [100, 100, 100] - After Differencing

   

**Stationarity**
A time series is **stationary** if its statistical properties (mean, variance, and autocorrelation) do not change over time.

  - Constant mean and variance over time.
  - No strong trend
  - No seasonality

**Why is Stationarity Important?**

- Many models (like ARIMA) require stationary data because it simplifies forecasting and ensures reliable results.

- Non-stationary data can lead to inaccurate predictions and misleading insights.

- Augmented Dickey-Fuller (ADF) Test: Tests if a series is stationary
  - If p-value < 0.05, reject H0(hypothesis), meaning the data is stationary.

  ```python
  from statsmodels.tsa.stattools import adfuller
  result = adfuller(time_series_data)
  print(f'ADF Statistic: {result[0]}')
  print(f'p-value: {result[1]}')
```


**3. Moving Average (MA):**  
- "Smooth Out the Noise"
- Uses dependency between an observation and residual errors from a moving average model applied to lagged observations. 
- MA learns from past mistakes to smooth out randomness. 

- **Parameter (q):** Determines the number of lagged forecast errors to include.  
- **Example:** 
   - Using past error terms to adjust future predictions. 
   - If your last guess was off by 10, adjust your next guess by considering that error.


**Putting It All Together**

ARIMA combines these three tools into one model to forecast time series:

- AR (p): How many past values to use?
- I (d): How many differences to make the data stationary?
- MA (q): How many past errors to use?

---

**When to Use ARIMA:**  
   - Data exhibits trends but lacks strong seasonality.  
   - Data is not stationary and requires differencing to remove trends.  
   - Need a reliable model for short-term forecasts. 

**Steps to Build an ARIMA Model:**  

   **1. Check for Stationarity:**  
   - Use visual inspection (line plots) and statistical tests like the Augmented Dickey-Fuller (ADF) test.  
   - **Stationarity Criteria:** Constant mean and variance over time. 

 

   **2. Make Data Stationary (Differencing):**  
   - Apply differencing to remove trends or seasonality.  

   **3. Identify ARIMA Parameters (p, d, q):**  
   - **ACF (Autocorrelation Function):** Helps determine the MA component (q).  
   - **PACF (Partial Autocorrelation Function):** Helps determine the AR component (p).  
   - **d (Integration):** Based on the number of differencing steps needed for stationarity.  

   **4. Fit the Model:**  
   - Use libraries like `statsmodels` in Python to fit the ARIMA model.  

   

**5. Evaluate the Model:**  
   - Use metrics like AIC (Akaike Information Criterion), BIC (Bayesian Information Criterion), and RMSE (Root Mean Squared Error).  

**Python Implementation** 

````python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from matplotlib import pyplot as plt

# Sample time series data
data = {
    "Date": pd.date_range(start="2024-01-01", end="2024-01-20"),
    "Value": [100, 102, 101, 104, 107, 110, 115, 113, 117, 119, 
              118, 120, 125, 130, 128, 135, 137, 140, 138, 142],
}

series = data['Value']

# Fit ARIMA model
model = ARIMA(series, order=(2, 1, 2))  # Example parameters (p, d, q)
model_fit = model.fit()

# Summary of the model
print(model_fit.summary())

# Forecasting
forecast = model_fit.forecast(steps=10)  # Predict the next 10 steps
print(forecast)

# Plot original data and forecast
plt.plot(series, label='Original')
plt.plot(forecast, label='Forecast', color='red')
plt.legend()
plt.show()
````

#### **Understanding the SARIMAX Results** (Optional - not all need to be understood)

1. **Dep. Variable (y):**  
   - The variable being predicted (dependent variable).

2. **No. Observations:**  
   - The total number of data points used in the model (here, 20 observations).

3. **Model:**  
   - The specific type of ARIMA model used, in this case, ARIMA(2, 1, 2), where:
     - 2 is the number of autoregressive terms (AR).
     - 1 is the number of differences (I) to make the data stationary.
     - 2 is the number of moving average terms (MA).

4. **Log Likelihood:**  
   - A measure of how well the model fits the data. Higher values indicate a better fit.

5. **AIC (Akaike Information Criterion):**  
   - A metric to evaluate the model's performance, where lower values indicate a better model.

6. **BIC (Bayesian Information Criterion):**  
   - Similar to AIC, but penalizes model complexity more. Lower values indicate a better model.

7. **HQIC (Hannan-Quinn Information Criterion):**  
   - Another criterion to assess model quality, balancing goodness-of-fit and model complexity.

8. **Covariance Type:**  
   - Refers to the method used for estimating the covariance of model parameters (here, "opg" stands for outer product of gradients).

9. **Coef (Coefficients):**  
   - The estimated values for each term in the model (AR and MA). These indicate the strength of the relationship between the variables and past values.

10. **Std Err (Standard Error):**  
    - A measure of the uncertainty or variability in the coefficient estimates.

11. **z:**  
    - The z-statistic, used to test the null hypothesis that a coefficient is zero. Higher values suggest that the coefficient is significant.

12. **P>|z| (P-Value):**  
    - The probability that the coefficient is not statistically significant. A value less than 0.05 indicates significance.

13. **[0.025, 0.975] (Confidence Interval):**  
    - The range within which the true coefficient value is likely to fall, with 95% confidence.

14. **sigma2 (Variance of Errors):**  
    - The variance of the residuals (errors). Smaller values indicate a better fit.

15. **Ljung-Box (L1) (Q):**  
    - A test for autocorrelation in the residuals (errors). A high p-value indicates no autocorrelation.

16. **Prob(Q):**  
    - The p-value from the Ljung-Box test. A value greater than 0.05 indicates no significant autocorrelation.

17. **Jarque-Bera (JB):**  
    - A test for normality of residuals. A high p-value suggests that residuals are normally distributed.

18. **Prob(JB):**  
    - The p-value for the Jarque-Bera test. Values greater than 0.05 indicate no significant deviation from normality.

19. **Heteroskedasticity (H):**  
    - A test for whether residuals have constant variance. A value greater than 1 suggests variance is not constant.

20. **Prob(H) (two-sided):**  
    - The p-value for the heteroskedasticity test. Values greater than 0.05 suggest no problem with heteroskedasticity.

21. **Skew:**  
    - A measure of the asymmetry of the residuals. Negative values indicate a left skew, positive values indicate a right skew.

22. **Kurtosis:**  
    - A measure of the "tailedness" of the residual distribution. Values around 3 indicate a normal distribution; higher or lower values indicate more or less extreme outliers, respectively.

    <h3>2. Exponential Smoothing
    </h3> </div>

Exponential smoothing is a time series forecasting method that uses weighted averages of past observations, with exponentially decreasing weights as observations get older.  


**Types of Exponential Smoothing Methods:**  


   **1. Simple Exponential Smoothing (SES):**  
   - **Suitable for:** Stationary data with no trend or seasonality.  
   - **How it works:** Assigns exponentially decreasing weights to older data points.  
   - **Key Parameter (Î±):**  
     - The smoothing factor (0 < Î± â‰¤ 1).  
     - Higher Î± gives more weight to recent observations.  
   - **Use Case:**  
     - Predicting daily demand for a product without a seasonal pattern.  



**Python Implementation** 

````python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Load data
series = [50, 52, 53, 54, 56, 57]  # Example data

# Fit model
model = SimpleExpSmoothing(series)
model_fit = model.fit(smoothing_level=0.2, optimized=False)

# Forecast
forecast = model_fit.forecast(steps=3)

print(forecast)
````

   **2. Holtâ€™s Linear Trend Model:**  
   - **Suitable for:** Data with a trend but no seasonality.  
   - **How it works:**  
     - Extends SES by introducing trend components.  
     - Separates the series into a base level and a trend.  
   - **Key Parameters:**  
     - Smoothing factors for level (Î±) and trend (Î²).  
   - **Use Case:**  
     - Forecasting sales growth for a product showing a linear increase.  

   

**Python Implementation** 

````python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Load data
series = [50, 52, 53, 54, 56, 57]  # Example data

# Fit model
model = ExponentialSmoothing(series, trend='add', seasonal=None)
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=3)
print(forecast)
````

**3. Holt-Winters Seasonal Model:**  
   - **Suitable for:** Data with both trend and seasonality.  
   - **How it works:**  
     - Adds a seasonal component to Holtâ€™s model.  
     - Supports both additive and multiplicative seasonality.  
   - **Key Parameters:**  
     - Smoothing factors for level (Î±), trend (Î²), and seasonality (Î³).  
   - **Use Case:**  
     - Forecasting monthly sales with seasonal fluctuations (e.g., holiday sales)

**Python Implementation** 

````python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Load data
series = [50, 52, 53, 54, 56, 57]  # Example data

# Fit model
model = ExponentialSmoothing(series, trend='add', seasonal='add', seasonal_periods=2)
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=4)

print(forecast)
````

    <h3>4. Hands-on Forecasting on Dataset
    </h3> </div>

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
````

````python
# Fabricated time series data
np.random.seed(42)
time_index = pd.date_range(start='2023-01-01', periods=60, freq='M')
trend = np.linspace(10, 20, 60)  # Linear trend
seasonality = 5 * np.sin(np.linspace(0, 3 * np.pi, 60))  # Seasonal pattern
noise = np.random.normal(0, 1, 60)  # Random noise


data = trend + seasonality + noise
````

````python
# Create a DataFrame
time_series_df = pd.DataFrame({'Date': time_index, 'Value': data})
time_series_df.set_index('Date', inplace=True)
````

````python
# Plot the fabricated data
plt.figure(figsize=(12, 6))
plt.plot(time_series_df, label='Fabricated Time Series')
plt.title('Fabricated Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
````

````python
# Decomposition
decompose_result = seasonal_decompose(time_series_df['Value'], model='additive', period=12)

# Plot decomposition
decompose_result.plot()
plt.show()
````

````python
# ARIMA model
# Differencing to achieve stationarity
time_series_diff = time_series_df['Value'].diff().dropna()

# Fitting ARIMA (1,1,1) for demonstration
arima_model = ARIMA(time_series_df['Value'], order=(1, 1, 1))
arima_result = arima_model.fit()
````

````python
# ARIMA forecast
arima_forecast = arima_result.forecast(steps=12)

# Simple Exponential Smoothing
ses_model = ExponentialSmoothing(time_series_df['Value'])
ses_fit = ses_model.fit(smoothing_level=0.2, optimized=False)
ses_forecast = ses_fit.forecast(steps=12)
````

````python
# Holt-Winters Model
hw_model = ExponentialSmoothing(time_series_df['Value'], trend='add', seasonal='add', seasonal_periods=12)
hw_fit = hw_model.fit()
hw_forecast = hw_fit.forecast(steps=12)

# Combine forecasts for visualization
forecast_df = pd.DataFrame({
    'ARIMA Forecast': arima_forecast,
    'SES Forecast': ses_forecast,
    'HW Forecast': hw_forecast
})
````

````python
# Plot forecasts
plt.figure(figsize=(12, 6))
plt.plot(time_series_df, label='Original Data')
plt.plot(arima_forecast, label='ARIMA Forecast', linestyle='--')
plt.plot(ses_forecast, label='SES Forecast', linestyle='--')
plt.plot(hw_forecast, label='Holt-Winters Forecast', linestyle='--')
plt.title('Time Series Forecasting Results')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

forecast_df
````
