<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3>1. Introduction to Data Warehousing
</h3> </div>

Data warehouse is a centralized repository designed to store large amounts of  data from multiple sources. 

Data warehouses help prepare data for data analytics, business intelligence (BI), data mining, machine learning (ML) and artificial intelligence (AI) initiatives.

An enterprise data warehouse (EDW) system enables an organization to run powerful analytics on large amounts of data (petabytes and more) in ways that a standard database cannot.


Features of Data Warehousing:

- **Integration**: Combines data from multiple sources into a unified view.
- **Historical Data**: Stores historical data for trend analysis.
- **Read-Optimized**: Designed for querying and reporting rather than transaction processing.
- **Query and Analysis**: Provides query analysis capabilities


- Data warehouses were born in the 1980s to optimize data analytics

- Business applications grew and new data sources exploded - including the World Wide Web, social media and Internet of Things (IoT).

- The need for larger storage and faster analysis grew.

- Legacy data warehouses could not manage new huge workloads. 

- In response, more flexible alternatives evolved, including cloud-native data warehouses and data lakehouses.



<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> OLTP vs. OLAP
</h4> </div>

##### **OLTP (Online Transaction Processing)**: (Capture data)

Focuses on managing transaction data in real-time (e.g., sales transactions). 
OLTP systems track large numbers of transactions, by large numbers of users, in real-time.

OLTP systems often use relational databases to record transactions such as:

- Banking and ATM transactions
- E-commerce and in-store purchases
- Hotel and airline reservations

##### **OLAP (Online Analytical Processing)**: (Analyze data)

Optimized for analysis and querying large datasets (e.g., sales reporting, trend analysis).
Common uses of OLAP include data mining and business intelligence apps. 



<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Data Warehouse Architecture
</h4> </div>


Data warehouses most often use a three-tier architecture.

##### **Bottom tier**

This tier is where the data flows from disparate data sources into a data warehouse server, where it is stored.
The data is most often moved through a process known as extract, transform, load (ETL) or sometimes a process known as extract, load, transform (ELT).






##### **Middle tier**

This tier is traditionally built around an analytics engine, most often an online analytical processing (OLAP) system that is designed to deliver fast analytics and query speeds.

##### **Top tier**

This tier includes a front-end user interface or reporting tool, which enables users to conduct ad hoc data analysis on their business data. Self-service business intelligence has multiple uses, such as the generation of reports based on historical data, identification of new opportunities or identifying process bottlenecks.

The architecture of a data warehouse typically consists of the following components:

#### 1.  **Data layer (or central database)**
The heart of the data warehouse and is supported by all other components. The data can be input from business applications, email lists, websites or any other relational databases. The data is physically stored on a server or set of servers.

#### 2. **ETL Process**
- **Extract**: Pull data from various sources (databases, APIs, flat files).
- **Transform**: Cleanse and format data for consistency.
- **Load**: Store the transformed data in the data warehouse.

#### 3. **Staging Area**
A temporary storage area where data is cleaned and transformed before being loaded into the data warehouse.

#### 4. **Data Mart**
A subset of the data warehouse that is focused on a specific business area (e.g., sales, finance).

#### 5. **Presentation Layer**
The interface for users to access and analyze the data through reporting tools and dashboards.

    <h4> Dimensional Modeling
</h4> </div>

Dimensional modeling is a design technique used for data warehouses. It involves creating a structure that is intuitive and easy to query.


### **Key Concepts**

- **Facts**: Quantitative data points, measurable items such as products sold and monetary sales amounts (e.g., sales amount, quantity sold).

- **Dimensions**: Descriptive attributes related to facts, reference information showing how the facts can be grouped and organized (e.g., dates of sale and industry sold to).


### **Schema Types**

In databases, a schema is the structure or organization of data. It defines how tables are related, including tables, fields, and their relationships (like primary and foreign keys). A star schema and snowflake schema are types of data modeling techniques, primarily used in data warehousing.

- **Star Schema:** A simpler schema with a central fact table linked directly to dimension tables.

- **Snowflake Schema:** A more normalized version of the star schema, where dimension tables are further split into related tables.

    <h4> Data Mining and Analytics
</h4> </div>

Data mining involves discovering patterns and insights from data stored in a data warehouse. Common techniques include:

- **Descriptive Analytics**: Understanding past data trends.
- **Predictive Analytics**: Using historical data to forecast future outcomes.
- **Prescriptive Analytics**: Suggesting actions based on data analysis.

### Tools and Technologies
- **SQL**: For querying and analyzing data.
- **Python/R**: For advanced analytics and data visualization.
- **Business Intelligence (BI) Tools**: Such as Tableau and Power BI for reporting.


<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3>Types of data warehouses
</h3> </di>

#### **Cloud data warehouse**
A cloud-based data warehouse is built to run in the cloud. It is often offered to organizations as a managed data-storage service in which the data warehouse infrastructure is managed by the cloud company. 

#### **On-premises or licensed data warehouse**

A business can purchase a data warehouse license and then deploy the warehouse as its own on-premises infrastructure. Although this is typically more expensive than a cloud data warehouse service, it might be a good choice for government entities, financial institutions or other organizations that want more control over their data or need to comply with strict security or data privacy standards.

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Database vs Data Warehouse vs Data Lake
</h4> </div>

### 1. **Database**
- **Purpose**: Used for storing and managing data for day-to-day operations.
- **Data type**: Structured data (organized in tables with rows and columns).
- **Speed**: Optimized for quick reads and writes.
- **Example**: Imagine a customer service app. It uses a database to store customer information, orders, and transactions, so agents can quickly look up and update details.
  
  **Example Technology**: PostgreSQL, MySQL, Oracle.

### 2. **Data Warehouse**
- **Purpose**: Stores large amounts of historical data for analysis and reporting.
- **Data type**: Structured data, often coming from multiple databases.
- **Speed**: Optimized for complex queries and data analysis.
- **Example**: A company might use a data warehouse to analyze sales trends over the past 5 years. The data comes from multiple sources (sales databases, marketing data, etc.) to give a full view.
  
  **Example Technology**: Amazon Redshift, Google BigQuery, Snowflake.

### 3. **Data Lake**
- **Purpose**: Stores vast amounts of raw data in its original form (structured, semi-structured, or unstructured).
- **Data type**: Any type of data, from text files to videos and sensor data.
- **Speed**: Less optimized for immediate use; used for big data processing, machine learning, etc.
- **Example**: A video streaming service like Netflix might use a data lake to store all kinds of dataâ€”movies, user viewing habits, app logsâ€”for future use in data analysis or AI models.
  
  **Example Technology**: AWS S3, Azure Data Lake, Hadoop.

#### **Simple Summary:**
- **Database** = Real-time operational data (e.g., customer orders).
- **Data Warehouse** = Historical data for reports and analysis (e.g., sales trends).
- **Data Lake** = Raw, mixed data for advanced processing (e.g., user behavior data).

### Objective: Create a Simple Data Warehouse Schemas

**Define the facts and dimensions** for simple sales reporting system:

   - **Fact Table**: Sales (with columns like sales_id, amount, product_id, time_id)
   
   - **Dimension Tables**: Product (product_id, name, category), Time (time_id, date, month, year)

- **Star Schema:** Simple structure, denormalized.
- **Snowflake Schema:** More normalized, where dimension tables are split into related tables.

#### **Star Schema Design**

**Step 1: Create a Fact Table for sales**

The fact table will store transactional data (e.g., sales).

```sql
CREATE TABLE fact_sales (
    sale_id INT PRIMARY KEY,
    product_id INT,
    customer_id INT,
    date_id INT,
    sales_amount DECIMAL(10, 2)
);
```

- **Fact Table:** `fact_sales` stores transactional information about sales.

**Step 2: Create Dimension Tables**

Each dimension table stores descriptive data related to the fact table.

- **Product Dimension Table:**

```sql
CREATE TABLE dim_product (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(50),
    category VARCHAR(50)
);
```

- **Customer Dimension Table:**

```sql
CREATE TABLE dim_customer (
    customer_id INT PRIMARY KEY,
    customer_name VARCHAR(50),
    region VARCHAR(50)
);
```

- **Date Dimension Table:**

```sql
CREATE TABLE dim_date (
    date_id INT PRIMARY KEY,
    date DATE,
    year INT,
    month INT
);
```

**Conclusion:**

In this star schema, `fact_sales` is the central table, and each dimension (`dim_product`, `dim_customer`, and `dim_date`) is connected directly to it. The relationships are defined by `product_id`, `customer_id`, and `date_id`.

**Step 2: Modify the Product Dimension Table**

Now, the `dim_product` table will reference the `dim_category` table.

```sql
CREATE TABLE dim_product (
    product_id INT PRIMARY KEY,
    product_name VARCHAR(50),
    category_id INT,
    FOREIGN KEY (category_id) REFERENCES dim_category(category_id)
);
```

```sql
-- Step 1: Add the category_id column to dim_product
ALTER TABLE dim_product
ADD COLUMN category_id INT;

-- Step 2: Add the foreign key constraint to link with dim_category
ALTER TABLE dim_product
ADD CONSTRAINT fk_category
FOREIGN KEY (category_id) REFERENCES dim_category(category_id);

**Conclusion**

Now, the `dim_product` table no longer stores category names directly but instead references the `dim_category` table, which makes the schema more normalized.

```yaml
               dim_category
                   |
               dim_product
                   |
dim_customer -- fact_sales -- dim_date
                   |
              dim_store


<div class="alert alert-block alert-warning">
    <b><font size="5"> Sample Entries for the hands-on activity</font> </b>
</div>

<!--- Mohammad Idrees Bhat | Tech Skills Trainer | AI/ML Consultant --->

`fact_sales`
```sql

INSERT INTO fact_sales (sale_id, product_id, customer_id, date_id, sales_amount) VALUES
(1, 101, 201, 301, 150.50),
(2, 102, 202, 302, 250.75),
(3, 103, 203, 303, 99.99);

`dim_product`
```sql
INSERT INTO dim_product (product_id, product_name, category_id) VALUES
(101, 'Laptop', 1),
(102, 'Smartphone', 2),
(103, 'Tablet', 3);

`dim_customer`
```sql

INSERT INTO dim_customer (customer_id, customer_name, region) VALUES
(201, 'John Doe', 'North America'),
(202, 'Jane Smith', 'Europe'),
(203, 'Alice Johnson', 'Asia');

`dim_date`
```sql

INSERT INTO dim_date (date_id, date, year, month) VALUES
(301, '2024-10-01', 2024, 10),
(302, '2024-10-02', 2024, 10),
(303, '2024-10-03', 2024, 10);

`dim_category`
```sql

INSERT INTO dim_category (category_id, category_name) VALUES
(1, 'Electronics'),
(2, 'Mobile Devices'),
(3, 'Tablets');

<div style="background-color: lightblue; color: white; padding: 10px; text-align: center;">
    <h1>_________________________________END________________________________
        <!--- Mohammad Idrees Bhat | Tech Skills Trainer | AI/ML Consultant --->
</h1> </div>

Some Additional Resources:

- https://www.ibm.com/topics/data-warehouse
- https://aws.amazon.com/what-is/data-warehouse/
- https://www.datacamp.com/tutorial/guide-to-data-warehousing-on-aws-with-redshift
- https://d1.awsstatic.com/training-and-certification/ramp-up_guides/Ramp-Up_Guide_Data_Analytics.pdf



<div style="background-color: #002147; color: #fff; padding: 30px; text-align: center;">
    <h1>THANK YOU!
        <!--- Mohammad Idrees Bhat | Tech Skills Trainer | AI/ML Consultant --->
</h1> </div>

<div class="alert alert-block alert-warning"  padding: 10px; text-align: center;">
    <font size="3"> Programming Interveiw Questions</font>
</div>

##### 1. What is a data warehouse, and how does it differ from a traditional database?
- **Follow-up:** Can you explain the key components of a data warehouse architecture?

---

##### 2. What is ETL, and what role does it play in data warehousing?
- **Follow-up:** Can you describe the ETL process and give examples of tools used for ETL?

---

##### 3. Explain the difference between a star schema and a snowflake schema in data warehousing.
- **Follow-up:** When would you choose one schema over the other?

---

##### 4. What is data normalization, and why is it important in the context of data warehousing?
- **Follow-up:** Can you give an example of how normalization can affect data retrieval performance?

---

##### 5. How do you ensure data quality and integrity in a data warehouse?
- **Follow-up:** What methods or tools do you use for data cleansing and validation during the ETL process?


### **SOLUTIONS**

# Data Warehousing Interview Questions and Answers

##### 1. What is a data warehouse, and how does it differ from a traditional database?
A data warehouse is a centralized repository that stores large volumes of historical data from multiple sources, optimized for analysis and reporting. It differs from a traditional database in that:
- **Purpose:** Databases are designed for transactional processing (CRUD operations), while data warehouses are designed for analytical processing (read-heavy operations).
- **Data Structure:** Data warehouses use denormalized structures (e.g., star and snowflake schemas) to optimize query performance, whereas databases typically use normalized structures to reduce data redundancy.
- **Data Source:** Data warehouses aggregate data from various operational systems, while databases usually serve a single application.

---

##### 2. What is ETL, and what role does it play in data warehousing?
ETL stands for Extract, Transform, Load. It is a process used to:
- **Extract** data from various source systems (e.g., databases, flat files).
- **Transform** the data into a suitable format for analysis (e.g., cleaning, aggregating, converting data types).
- **Load** the transformed data into the data warehouse.

ETL is crucial because it prepares and consolidates data, ensuring it is accurate and reliable for reporting and analysis.

---

##### 3. Explain the difference between a star schema and a snowflake schema in data warehousing.
- **Star Schema:** In a star schema, a central fact table is surrounded by dimension tables. The structure is denormalized, which simplifies queries and improves performance.
  
- **Snowflake Schema:** A snowflake schema is a more normalized version of the star schema, where dimension tables are further broken down into sub-dimensions. This can reduce data redundancy but may complicate queries and impact performance.

**When to choose one over the other:**
- Choose a star schema for simpler queries and better performance.
- Choose a snowflake schema for better data integrity and when storage space is a concern.

---

##### 4. What is data normalization, and why is it important in the context of data warehousing?
Data normalization is the process of organizing data in a database to minimize redundancy and dependency. In data warehousing, normalization is important because:
- It improves data integrity by reducing anomalies during data operations.
- It can simplify the structure of the data model, making it easier to maintain.

However, in the context of data warehousing, too much normalization can lead to complex queries and slower performance, so a balance is often needed.

---

##### 5. How do you ensure data quality and integrity in a data warehouse?
To ensure data quality and integrity, you can:
- **Implement Data Validation Rules:** Use validation rules during the ETL process to check for accuracy and completeness.
- **Use Data Profiling Tools:** Analyze data to understand its quality and identify issues before loading it into the warehouse.
- **Conduct Regular Audits:** Perform data quality audits to check for inconsistencies or anomalies.

**Methods and tools used:**
- Tools like Apache Nifi, Talend, and Informatica can assist in data cleansing and validation during the ETL process.


<!--- Mohammad Idrees Bhat | Mohammad Idrees Bhat --->

<h2 style="background-color: #ffe4e1; color: #2f4f4f; padding: 10px; border-radius: 10px; width: 350px; text-align: center; float: right; margin: 20px 0;">
    Mohammad Idrees Bhat<br>
    <span style="font-size: 12px; color: #696969;">
        Tech Skills Trainer | AI/ML Consultant
    </span>
</h2>

<!--- Mohammad Idrees Bhat | Tech Skills Trainer | AI/ML Consultant --->
